### Title

```python
import matplotlib.pyplot as plt

month = ["Jan", "Feb", "Mar", "Apr"]
sales = [100,150,200,300]

plt.plot(month, sales)
plt.title("Monthly sales trend", fontsize=16, color='blue', fontweight='bold')
plt.show()
```

### Axis labels

```python
import matplotlib.pyplot as plt

month = ["Jan", "Feb", "Mar", "Apr"]
sales = [100,150,200,300]

plt.plot(month, sales)
plt.title("Monthly sales trend", fontsize=16, color='blue', fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Sales trend")
plt.show()
```

### Legends

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]

sales_2025 = [10,20,30,40]
sales_2026 = [15,25,35,45]

plt.plot(x, sales_2025, label= "2025")
plt.plot(x, sales_2026, label = "2026")

plt.title("Sales trend")

plt.legend(loc="upper left")

plt.show()
```

### Gridlines

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]

sales_2025 = [10,20,30,40]
sales_2026 = [15,25,35,45]

plt.plot(x, sales_2025, label= "2025")
plt.plot(x, sales_2026, label = "2026")

plt.title("Sales trend")

plt.legend(loc="upper left")
plt.grid(color="blue", linestyle="--", linewidth=0.5)

plt.show()
```

### Axis Limit

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3], [20,50,80])

plt.ylim(0,100)
plt.xlim(0,10)

plt.show()
```

### Tick customization

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3], [20,50,80])

plt.xticks([1,2,3], ["Jan", "Feb", "Mar"], rotation = 90)

plt.show()
```

### Figure Size

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
plt.plot([1,2,3], [20,50,80])

plt.xticks([1,2,3], ["Jan", "Feb", "Mar"], rotation = 90)

plt.show()
```

### Background style / Theme

```python
import matplotlib.pyplot as plt

print(plt.style.available)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
plt.plot([1,2,3], [20,50,80])

plt.xticks([1,2,3], ["Jan", "Feb", "Mar"], rotation = 90)
plt.style.use("dark_background")

plt.show()
```

### Saving Charts

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
plt.plot([1,2,3], [20,50,80])

plt.xticks([1,2,3], ["Jan", "Feb", "Mar"], rotation = 90)
plt.style.use("dark_background")

plt.savefig("sales_chart.png")
plt.savefig("sales_chart.pdf")

plt.show()
```
