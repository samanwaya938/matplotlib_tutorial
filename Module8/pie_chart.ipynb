```python
import matplotlib.pyplot as plt

data = [40,30,20,10]
labels = ["A","B","C","D"]
color = ["red","blue",'green','orange']
explode = [0.1,0,0.1,0]

plt.pie(data, labels=labels, autopct='%1.1f%%', colors=color, explode=explode, startangle=90, shadow=True)
plt.show()
```

### Object oriented approch

```python
import matplotlib.pyplot as plt

data = [40,30,20,10]
labels = ["A","B","C","D"]
color = ["red","blue",'green','orange']
explode = [0.1,0,0.1,0]

fig, ax = plt.subplots()

ax.pie(data, labels=labels, autopct='%1.1f%%', colors=color, explode=explode, startangle=90, shadow=True)
plt.show()
```
