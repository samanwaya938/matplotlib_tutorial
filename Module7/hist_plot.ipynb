```python
import matplotlib.pyplot as plt

marks = [45,50,52,55,60,62,65,70,72,80,85,90]

plt.hist(marks, color="green", edgecolor="black", density=True, bins=[40,50,60,70,80,90])
plt.title("Distribution of student marks")
plt.xlabel("marks range")
plt.ylabel("Frequency")
plt.show()
```

### Multiple Histogram

```python
import matplotlib.pyplot as plt

marks_classA = [45,50,55,60,65,70]
marks_classB = [50,55,60,75,80,85]

plt.hist(marks_classA, color='red', alpha=0.5, label="Class A")
plt.hist(marks_classB, color='green', alpha=0.5, label="Class B")

plt.legend()
plt.show()
```

### Object Oriented approch

```python
import matplotlib.pyplot as plt

marks = [45,50,52,55,60,62,65,70,72,80,85,90]

fig, ax = plt.subplots()

ax.hist(marks, color="green", edgecolor="black", density=True, bins=[40,50,60,70,80,90])
plt.title("Distribution of student marks")
plt.xlabel("marks range")
plt.ylabel("Frequency")
plt.show()
```
