```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,15,25,30]
colors= [100,200,300,400,500]

plt.scatter(x,y, c=colors, s=100, cmap='viridis')
plt.title("Height vs Weight")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

### Multiple scatter group

```python
import matplotlib.pyplot as plt

x1=[1,2,3]
y1=[10,20,30]

x2=[1,2,3]
y2=[30,20,10]

plt.scatter(x1,y1, color='blue', label="Group 1")
plt.scatter(x2,y2, color='red', label="Group 2")

plt.legend()
plt.show()
```

### Correlation

### Positive

```python
import matplotlib.pyplot as plt

study_hour = [1,2,3,4,5]
marks = [40,50,60,70,80]

plt.scatter(study_hour,marks)
plt.title("Positive Correlation")
plt.xlabel("Study hour")
plt.ylabel("Marks")
plt.show()
```

### Negetive Correlation

```python
import matplotlib.pyplot as plt

exc_hour = [1,2,3,4,5]
body_fat = [35,30,25,20,15]

plt.scatter(exc_hour,body_fat)
plt.title("Negetive Correlation")
plt.xlabel("Execise hour")
plt.ylabel("Body fat")
plt.show()
```

### No correlation

```python
import matplotlib.pyplot as plt

shoe_size = [6,7,8,9,10]
marks = [80,45,90,60,72]

plt.scatter(shoe_size,marks)
plt.title("No Correlation")
plt.xlabel("Shoe size")
plt.ylabel("Marks")
plt.show()
```

### Object Oriented Approch

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,15,25,30]
colors= [100,200,300,400,500]

fig, ax = plt.subplots()

ax.scatter(x,y, c=colors, s=100, cmap='viridis')
plt.title("Height vs Weight")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```
