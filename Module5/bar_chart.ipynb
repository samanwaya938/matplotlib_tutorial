```python
import matplotlib.pyplot as plt

colors = ['red','blue', 'green', 'purple']

categories = ["A","B","C","D"]
values = [10,20,15,25]

plt.barh(categories,values, color = colors, height=0.5)
plt.title("Sales by category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

### Multiple bars (Grouped bar chart)

```python
import matplotlib.pylab as plt
import numpy as np

x = np.arange(4) # 0,1,2,3

sales_2025 = [10,20,15,25]
sales_2026 = [15,25,20,30]

plt.bar(x-0.2, sales_2025, width=0.4, label= '2025')
plt.bar(x+0.2, sales_2026, width=0.4, label= '2026')

plt.xticks(x, ["A", "B", "C", "D"])
plt.legend()
plt.show()
```

```python
import matplotlib.pylab as plt
import numpy as np

x = np.arange(4) # 0,1,2,3

sales_2025 = [10,20,15,25]
sales_2026 = [15,25,20,30]

fig, ax = plt.subplots()

ax.bar(x-0.2, sales_2025, width=0.4, label= '2025')
ax.bar(x+0.2, sales_2026, width=0.4, label= '2026')

plt.xticks(x, ["A", "B", "C", "D"])
ax.legend()
plt.show()
```
