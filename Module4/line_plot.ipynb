```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,30,40]

plt.plot(x,y, color= 'red', linestyle= '--', marker='o')
plt.title("Sales Growth")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y1 = [10,20,30,40]
y2 = [40,30,20,10]

plt.plot(x,y1, color= 'red', linestyle= '--', marker='o', label="Growth")
plt.plot(x,y2, color= 'blue', linestyle= '--', marker='o', label="Decline")

plt.title("Sales Growth")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
