### Subplot (Singular)

```python
## 2*2 matrix

import matplotlib.pyplot as plt

plt.subplot(2,2,1)
plt.plot([1,2,3],[4,5,6])

plt.subplot(2,2,2)
plt.plot([1,2,3],[6,5,4])

plt.subplot(2,2,3)
plt.plot([1,2,3],[2,3,4])

plt.subplot(2,2,4)
plt.plot([1,2,3],[9,8,7])

plt.show()
```

### Subplots (Plural)

### 1 row & 3 column (1D array)

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,3)

ax[0].plot([1,2,3],[4,5,6])
ax[1].plot([1,2,3],[6,5,4])
ax[2].plot([1,2,3],[9,8,7])

plt.show()
```

### 3 row & 1 column (1D array)

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(3,1)

ax[0].plot([1,2,3],[4,5,6])
ax[1].plot([1,2,3],[6,5,4])
ax[2].plot([1,2,3],[9,8,7])

plt.show()
```

### 2*2 matrix (2D array)

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2,2, figsize=(10,5))

ax[0,0].plot([1,2,3],[4,5,6])
ax[0,1].plot([1,2,3],[6,5,4])
ax[1,0].plot([1,2,3],[9,8,7])
ax[1,1].plot([1,2,3],[2,3,4])

plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2,2, figsize=(10,5))
ax = ax.flatten()

ax[0].plot([1,2,3],[4,5,6])
ax[1].plot([1,2,3],[6,5,4])
ax[2].plot([1,2,3],[9,8,7])
ax[3].plot([1,2,3],[2,3,4])

plt.show()
```

### 3*4 (2D Array)

```python
import matplotlib.pylab as plt

fig, ax = plt.subplots(3,4, figsize = (12,8))

ax[0,0].plot([1,2,3],[4,5,6])
ax[0,1].plot([1,2,3],[6,5,4])
ax[0,2].plot([1,2,3],[9,8,7])
ax[0,3].plot([1,2,3],[2,3,4])

ax[1,0].plot([1,2,3],[4,5,6])
ax[1,1].plot([1,2,3],[6,5,4])
ax[1,2].plot([1,2,3],[9,8,7])
ax[1,3].plot([1,2,3],[2,3,4])

ax[2,0].plot([1,2,3],[4,5,6])
ax[2,1].plot([1,2,3],[6,5,4])
ax[2,2].plot([1,2,3],[9,8,7])
ax[2,3].plot([1,2,3],[2,3,4])

plt.show()
```

```python
import matplotlib.pylab as plt

fig, ax = plt.subplots(3,4, figsize = (12,8))

count = 1

for i in range(3):   # rows
  for j in range(4):  # columns
    ax[i,j].plot([1,2,3], [count, count+1, count+2])
    count +=1

plt.show()
```
