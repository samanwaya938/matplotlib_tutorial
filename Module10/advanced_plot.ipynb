### Box plot

```python
import matplotlib.pyplot as plt

data = [10,20,30,40,50,100]

plt.boxplot(data)
plt.title("Box plot Example")
plt.show()
```

### Multi box plot

```python
data1 = [10,20,30,40]
data2 = [20,30,40,50]

plt.boxplot([data1,data2])
plt.show()
```

### Violin plot ( box plot + distribution shape)

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(0,1, 100)

plt.violinplot(data)
plt.show()

```

### Area plot

```python
x= [1,2,3,4]
y=[10,20,15,30]

plt.fill_between(x,y)
plt.show()
```

### Stack plot

```python
x = [1,2,3,4]

product_a = [10,20,30,40]
product_b = [5,10,15,20]

plt.stackplot(x, product_a,product_b, labels=["A","B"])
plt.legend()
plt.show()
```

### Error bar plot

```python
x = [1,2,3,4]
y = [10,20,30,40]

error = [2,3,4,2]

plt.errorbar(x,y, yerr=error)
plt.show()
```

### Stem plot (dot + vertical line)

```python
x = [1,2,3,4]
y = [10,20,30,40]

plt.stem(x,y)
plt.show()
```

### Step plot

```python
x = [1,2,3,4]
y = [10,20,20,30]

plt.step(x,y)
plt.show()
```
