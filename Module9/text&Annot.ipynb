### Text

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,30,40]

plt.plot(x,y)
plt.text(2,20, "Mid Point", fontsize=12, color='red', fontweight='bold')
plt.show()
```

### Annotation

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,30,40]

plt.plot(x,y)
plt.annotate("Peak sales", xy=(4,40), arrowprops=dict(arrowstyle='->', color='red'), xytext=(3,50))
plt.show()
```

```python
import matplotlib.pyplot as plt

sales = [100,150,120,200]
months = ["Jan", "Feb", "Mar", "Apr"]

max_value = max(sales)
max_index = sales.index(max_value)

min_value = min(sales)
min_index = sales.index(min_value)

plt.annotate(
  "Highest Sales",
  xy=(months[max_index], max_value),
  xytext=(months[max_index], max_value+30),
  arrowprops=dict(arrowstyle="->", color="green")
)

plt.annotate(
  "Lowest Sales",
  xy=(months[min_index], min_value),
  xytext=(months[min_index], min_value - 40),
  arrowprops=dict(arrowstyle="->", color="red")
)

plt.plot(months, sales, marker="o")
plt.title("Monthly sales analysis")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```
